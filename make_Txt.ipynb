```python
import json
import re
import subprocess

# Baca file notebook
notebook_name = "Scrappting.ipynb"
with open(notebook_name, "r", encoding='utf-8') as f:
    notebook = json.load(f)

# Kumpulkan package dari import dan pip install
imports = set()
pip_packages = {}

for cell in notebook["cells"]:
    if cell["cell_type"] == "code":
        for line in cell["source"]:
            line = line.strip()
            
            # Skip komentar
            if line.startswith("#"):
                continue
            
            # Tangkap import
            if line.startswith("import "):
                libs = re.split(r"import | as |,", line)
                for lib in libs[1:]:
                    if lib.strip():
                        imports.add(lib.strip().split(".")[0])
            elif line.startswith("from "):
                lib = line.split(" ")[1].split(".")[0]
                imports.add(lib)
            
            # Tangkap !pip install atau %pip install
            if "pip install" in line:
                parts = re.split(r"pip install |--", line)
                packages = parts[1].split() if len(parts) > 1 else []
                for pkg in packages:
                    if not pkg.startswith("-"):
                        pkg_info = re.split(r"==|>=|<=|>|<", pkg)
                        pkg_name = pkg_info[0]
                        pkg_version = pkg_info[1] if len(pkg_info) > 1 else None
                        pip_packages[pkg_name] = pkg_version

# Gabungkan hasil
all_packages = sorted(imports.union(pip_packages.keys()))

# Tulis ke requirements.txt
with open("requirements.txt", "w") as f:
    for pkg in all_packages:
        if pkg in pip_packages and pip_packages[pkg]:
            f.write(f"{pkg}=={pip_packages[pkg]}\n")
        else:
            # Jika versi tidak diketahui, coba dapatkan versi yang terinstal
            try:
                version = subprocess.check_output(["pip", "show", pkg], text=True)
                version = re.search(r"Version: (.*)", version).group(1)
                f.write(f"{pkg}=={version}\n")
            except:
                f.write(f"{pkg}\n")

print("requirements.txt berhasil dibuat!")
```
