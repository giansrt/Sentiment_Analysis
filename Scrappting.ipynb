```python
from google_play_scraper import app, Sort, reviews
import pandas as pd
import time

def mass_scraper(package_names, num_reviews=10000):
    all_reviews = []

    for package in package_names:
        continuation_token = None
        batch_size = 100  # Max per batch

        for _ in range(num_reviews // batch_size):
            try:
                result, continuation_token = reviews(
                    package,
                    lang='id',
                    country='id',
                    sort=Sort.NEWEST,
                    count=batch_size,
                    continuation_token=continuation_token
                )
                all_reviews.extend(result)

                # Delay untuk menghindari blokir
                time.sleep(2)

                if not continuation_token:
                    break

            except Exception as e:
                print(f"Error: {e}")
                break

    return pd.DataFrame(all_reviews)

apps = [
    # 'com.maxridemobile.app',  # Contoh package name untuk Maxride
    'com.grabtaxi.passenger', # Package name resmi untuk Grab
    # 'com.gojek.app'           # Package name resmi untuk Gojek
]
```

```python
# Gunakan fungsi yang diperbarui
df = mass_scraper(apps, 10000)
```

```python
df.to_csv('reviews.csv', index=False)
df.head()
```
